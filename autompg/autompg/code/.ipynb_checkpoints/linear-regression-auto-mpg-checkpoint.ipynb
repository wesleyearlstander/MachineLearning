```python
data = pd.read_csv('../data/auto-mpg.csv')
data.head()
```

```python
# mpg is target value
y = data[['mpg']].values
# we only use displacement as feature
x = data[['displacement']].values
x = []
y = np.vstack(np.ones((360,1)))
for i in range(360):
    x.append([1])
    x[i].append(i * np.pi/180) 
    x[i].append(x[i][1]**2)
    x[i].append(x[i][1]**3)
    x[i].append(x[i][1]**4)
    y[i] = np.around(np.cos(x[i][1]),2)
    
                
print(x, y)
# split data into train and test sets with (aproximately 50% test splits)
index_train = []
index_test = []
for i in np.arange(y.shape[0]):
    if ((i%2)==0):
        index_train.append(i)
    else:
        index_test.append(i)
x_train = [x[i] for i in index_train] 
y_train = y[index_train]
x_test = [x[i] for i in index_test] 
y_test = y[index_test]
pl.figure()
pl.scatter([x_train[i][1] for i in range(len(x_train))] , y_train[:,0], label='train data')
pl.scatter([x_test[i][1] for i in range(len(x_test))], y_test[:,0], label='test data')
pl.xlabel('displacement')
pl.ylabel('mpg')
pl.legend();
```

# Normalize features

```python
print('x train dataset before normalization')
print(x_train[0:10])
print('x test dataset before normalization')
print(x_test[0:10])
```

```python
mean_vector_train = np.mean(x_train, axis=0)
std_vector_train = np.std(x_train, axis=0)
x_train = (x_train - mean_vector_train) / (std_vector_train + 0.0000000000001)
x_test = (x_test - mean_vector_train) / (std_vector_train + 0.0000000000001)
```

```python
print('x train dataset after normalization')
print(x_train[0:10])
print('x test dataset after normalization')
print(x_test[0:10])
```

```python
pl.figure()
pl.scatter([x_train[i][1] for i in range(len(x_train))], y_train[:,0], label='train data (after normalization)')
pl.scatter([x_test[i][1] for i in range(len(x_test))], y_test[:,0], label='test data (after normalization)')
pl.xlabel('displacement')
pl.ylabel('mpg')
pl.legend();
```

# Create augmented feature vector

```python
x_train_aug = np.hstack((np.ones((x_train.shape[0],1)), x_train))
x_test_aug = np.hstack((np.ones((x_test.shape[0],1)), x_test))
print(x_train_aug)
```

```python
print('augmented x train')
print(x_train_aug[0:10])
print('augmented x test')
print(x_test_aug[0:10])
```

# Define linear regression model

```python
def f(x, theta):
    y = np.dot(x, theta)
    return y
```

# Batch gradient descent learning

```python
def learn(x, y, theta, h, alpha = 0.1, epsilon=0.001, t_max=20):
    t = 1
    theta_trajectory = [theta]
    while True:
        e = (h(x, theta) - y)
        x_e = (x * e)
        theta_ = theta.copy()
        theta = theta - alpha * np.mean(x_e, axis=0)[:,np.newaxis]
        theta_trajectory.append(theta)
        if t > t_max:
            break
        t = t + 1
    return theta, theta_trajectory
```

# Learn the parameters of the model

```python
theta_initial = np.zeros((x_train_aug.shape[1],1))
#print((f(x_train, theta_initial) - y_train))
theta_learned, theta_trajectory = learn(x_train_aug, y_train, theta_initial, f)
print(theta_learned)
```

# Plot the trajectory of the model parameters during the learning process

```python
pl.figure()
for i in np.arange(len(theta_trajectory)):
    pl.scatter(theta_trajectory[i][0,0], theta_trajectory[i][1,0], color='blue')
    if i < len(theta_trajectory)-1:
        x = [theta_trajectory[i][0,0], theta_trajectory[i+1][0,0]]
        y = [theta_trajectory[i][1,0], theta_trajectory[i+1][1,0]]
        pl.plot(x,y,color='red')
        pl.text(theta_trajectory[i][0,0], theta_trajectory[i][1,0], r'$\Theta^{{({:d})}}$'.format(i))
pl.xlabel(r'$\theta_{0}$')
pl.ylabel(r'$\theta_{1}$');
```

# Test predictions on train data

```python
y_train_predicted = f(x_train_aug, theta_learned)
print(y_train_predicted)
pl.figure()
pl.scatter(x_train[:,1], y_train[:,0], label='train data', c='blue')
pl.scatter(x_train[:,1], y_train_predicted[:,0], label='prediction', c='red')
pl.xlabel('displacement')
pl.ylabel('mpg')
pl.legend();
```

# Test predictions on test data

```python
y_test_predicted = f(x_test_aug, theta_learned)
pl.figure()
pl.scatter(x_test_aug[:,1], y_test[:,0], label='train data', c='green')
pl.scatter(x_test_aug[:,1], y_test_predicted[:,0], label='prediction', c='red')
pl.xlabel('displacement')
pl.ylabel('mpg')
pl.legend();
```
